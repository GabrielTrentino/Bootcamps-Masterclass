```python
# Importando as bibliotecas:
import cv2
import requests
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Baixando a imagem do GitHub:
DownURL = 'https://github.com/GabrielTrentino/Bootcamps-Masterclass/blob/main/02%20-%20Masterclass/02.1%20-%20SIGMOIDAL%20-%20Vis%C3%A3o%20Computacional/Imagens/placa-carro.png?raw=true'
img_data = requests.get(DownURL).content
with open('placa-carro.png', 'wb') as handler:
    handler.write(img_data)

# Lendo a imagem:
img = cv2.imread('placa-carro.png')

# Redimensionando a imagem:
img = cv2.resize(img, None, fx=0.8, fy=0.8, interpolation=cv2.INTER_LINEAR)
```

```python
cv2_imshow(img)
```
